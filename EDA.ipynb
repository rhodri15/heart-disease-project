## Defining Useful Functions

```python
def get_histogram(data, variable, bins):
    """Function to generate a histogram

    Args:
        data (pandas.DataFrame): dataframe
        variable (str): column from the dataframe
        bins (int): number of desired bins
    """
    plt.figure(figsize = (8,5))
    plt.hist(data[variable], bins, edgecolor='black')
    plt.title('Histogram of the ' + variable + ' variable')
    plt.ylabel('count')
    plt.xlabel(variable)
    plt.show()
```

```python
def get_boxplot(data, variable):
    """Function to generate a boxplot.

    Args:
        data (pandas.DataFrame): dataframe
        variable (str): column from the dataframe
    """
    plt.boxplot(data[variable])
    plt.title('Boxplot of the ' + variable + ' variable')
    plt.ylabel(variable)
    plt.show()
```

```python
def get_correlation_heatmap(corr_df):
    """Function to generate a correlation heatmap.

    Args:
        corr_df (pandas.DataFrame.corr()): correlation matrix dataframe
    """
    axis_labels = list(corr_df.columns)
    #y_axis_labels = list(reversed(x_axis_labels))
    
    fig, ax = plt.subplots()
    
    #ax.xaxis.tick_top() (make horizontalalignment='left')
    ax.set_xticks([x + 0.5 for x in range(len(corr_df))])
    ax.set_xticklabels(axis_labels, rotation=45, horizontalalignment='right')
    ax.set_yticks([x + 0.5 for x in range(len(corr_df))])
    ax.set_yticklabels(axis_labels)
    
    pcm = ax.pcolor(corr_df, cmap='RdYlBu')
    fig.colorbar(mappable=pcm)
    
    ax.set_title('Correlation Heatmap', fontsize=15)
    plt.show()
```

```python
def mean(data, variable):
    """
    Function to find the mean of a variable
    """
    sum = data[variable].sum()
    n = len(data)
    x_bar = sum/n
    return x_bar
```
