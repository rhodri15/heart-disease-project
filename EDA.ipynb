## Defining Useful Functions

```python
def get_histogram(data, variable, bins):
    """
    Function to show a histogram plot
    """
    plt.figure(figsize = (8,5))
    plt.hist(data[variable], bins, edgecolor='black')
    plt.title('Histogram of the ' + variable + ' variable')
    plt.ylabel('count')
    plt.xlabel(variable)
    return plt.show
```

```python
def get_boxplot(data, variable):
    plt.boxplot(data[variable])
    plt.title('Boxplot of the ' + variable + ' variable')
    plt.ylabel(variable)
    return plt.show
```

```python
def get_correlation_heatmap(corr_df):
    x_axis_labels = list(corr_df.columns)
    y_axis_labels = list(reversed(x_axis_labels))
    
    fig, ax = plt.subplots()
    
    ax.xaxis.tick_top()
    ax.set_xticks([x + 0.5 for x in range(len(corr_df))])
    ax.set_xticklabels(x_axis_labels, rotation=45, horizontalalignment='left')
    ax.set_yticks([x + 0.5 for x in range(len(corr_df))])
    ax.set_yticklabels(y_axis_labels)
    
    pcm = ax.pcolormesh(corr_df, cmap='RdYlBu')
    fig.colorbar(mappable=pcm, ax=ax)
    
    ax.set_title('Correlation Heatmap', fontsize=15)
    fig.show()
    
```

```python
def mean(data, variable):
    """
    Function to find the mean of a variable
    """
    sum = data[variable].sum()
    n = len(data)
    x_bar = sum/n
    return x_bar
```
