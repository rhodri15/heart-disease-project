## Defining Useful Functions

```python
def get_histogram(data, variable, bins):
    """
    Function to show a histogram plot
    """
    plt.figure(figsize = (8,5))
    plt.hist(data[variable], bins, edgecolor='black')
    plt.title('Histogram of the ' + variable + ' variable')
    plt.ylabel('count')
    plt.xlabel(variable)
    return plt.show
```

```python
def get_boxplot(data, variable):
    plt.boxplot(data[variable])
    plt.title('Boxplot of the ' + variable + ' variable')
    plt.ylabel(variable)
    return plt.show
```

```python
def mean(data, variable):
    """
    Function to find the mean of a variable
    """
    sum = data[variable].sum()
    n = data.shape[0]
    x_bar = sum/n
    return x_bar
```
